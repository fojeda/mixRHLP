```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_graph(graph):
    plt.figure(figsize=(5, 3))
    nx.draw_spectral(
        graph,
        with_labels=True,
        node_size=400,
        font_color='white'
        )
```

## Chain üîó

```python
# Chain: A -> B -> C
graph = nx.DiGraph()

# Define edges
graph.add_edges_from(
    [
        ('A', 'B'),
        ('B', 'C')
    ]
)
```

```python
# Visualize
draw_graph(graph)
```
